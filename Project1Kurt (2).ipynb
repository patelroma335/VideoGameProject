```python
#By Year Total Profits of each Region
yearlydata=vgdata.groupby(["Year_of_Release"])

#yearly total profits for each region
na_total5=yearlydata["NA_Sales"].sum()
jp_total5=yearlydata["JP_Sales"].sum()
eu_total5=yearlydata["EU_Sales"].sum()
o_total5=yearlydata["Other_Sales"].sum()
g_total5=yearlydata["Global_Sales"].sum()


#By year regional shooter profits
yearly_shooter1=vgdata[vgdata["Genre"].isin(["Shooter"])]
yearly_shooter=yearly_shooter1.groupby(["Year_of_Release"])

#yearly shooter profits
na_shooter=yearly_shooter["NA_Sales"].sum()
jp_shooter=yearly_shooter["JP_Sales"].sum()
eu_shooter=yearly_shooter["EU_Sales"].sum()
o_shooter=yearly_shooter["Other_Sales"].sum()
g_shooter=yearly_shooter["Global_Sales"].sum()

#yearly RP Profits
yearly_rp1=vgdata[vgdata["Genre"].isin(["Role-Playing"])]
yearly_rp=yearly_rp1.groupby(["Year_of_Release"])

na_rp=yearly_rp["NA_Sales"].sum()
jp_rp=yearly_rp["JP_Sales"].sum()
eu_rp=yearly_rp["EU_Sales"].sum()
o_rp=yearly_rp["Other_Sales"].sum()
g_rp=yearly_rp["Global_Sales"].sum()

#Add Sports as intermediate
yearly_sports1=vgdata[vgdata["Genre"].isin(["Sports"])]
yearly_sports=yearly_sports1.groupby(["Year_of_Release"])

na_sports=yearly_sports["NA_Sales"].sum()
jp_sports=yearly_sports["JP_Sales"].sum()
eu_sports=yearly_sports["EU_Sales"].sum()
o_sports=yearly_sports["Other_Sales"].sum()
g_sports=yearly_sports["Global_Sales"].sum()


yearly_comp=pd.DataFrame({"NA Shooter market share by year": na_shooter/na_total5,
                     "NA RP market share by year": na_rp/na_total5,
                        #"NA Sports market share by year": na_sports/na_total5,
                    "EU Shooter market share by year": eu_shooter/eu_total5,
                     "EU RP market share by year": eu_rp/eu_total5,
                        #"EU Sports market share by year": eu_sports/na_total5,
                    "JP Shooter market share by year": jp_shooter/jp_total5,
                     "JP RP market share by year": jp_rp/jp_total5,
                        #"JP Sports market share by year": jp_sports/na_total5,
                    "Other Shooter market share by year": o_shooter/o_total5,
                    "Other RP market share by year": o_rp/o_total5})
                        #"Other Sports market share by year": o_sports/na_total5,})

#Drop Null values in early years
#yearly_comp.dropna(how="any")
yearly_comp.dropna(subset=['NA RP market share by year'], inplace=True)
yearly_comp.head(10)
```

```python
#sum_df.index.name="Genre"
yearly_comp.plot(kind="line", figsize=(20,10))
plt.title("Yearly Market Share for RP and Shooters per Region")
plt.savefig("Yearly Market Share for RP and Shooter per Region")
plt.show
```

```python
#Yearly Profits by Region
yearly_profits=pd.DataFrame({"NA Shooter Sales by year": na_shooter,
                     "NA RP Sales by year": na_rp,
                    "EU Shooter Sales by year": eu_shooter,
                     "EU RP Sales by year": eu_rp,
                    "JP Shooter Sales by year": jp_shooter,
                     "JP RP Sales by year": jp_rp,
                    "Other Sales by year": o_shooter,
                    "Other RP Sales by year": o_rp,})

#Drop Null values in early years
#yearly_comp.dropna(how="any")
yearly_profits.dropna(subset=['NA RP Sales by year'], inplace=True)
yearly_profits.head(10)
```

```python
#Yearly Total Profits for Shooter, RP
yearly_profits.plot(kind="line", figsize=(15,10))
plt.title("Yearly Total Sales for Shooters, RP Video Games by Region")
plt.savefig("Yearly Total Sales for SHooter and RP")
plt.show
```

```python
#Plot Global Rating Sales
ratingdata1=vgdata[vgdata["Rating"].isin(["E"])]
rate_E1=ratingdata1.groupby(["Year_of_Release"])
rate_E=rate_E1["Global_Sales"].sum()

ratingdata2=vgdata[vgdata["Rating"].isin(["E10+"])]
rate_Eten1=ratingdata2.groupby(["Year_of_Release"])
rate_Eten=rate_Eten1["Global_Sales"].sum()

ratingdata3=vgdata[vgdata["Rating"].isin(["T"])]
rate_T1=ratingdata3.groupby(["Year_of_Release"])
rate_T=rate_T1["Global_Sales"].sum()

ratingdata4=vgdata[vgdata["Rating"].isin(["M"])]
rate_M1=ratingdata4.groupby(["Year_of_Release"])
rate_M=rate_M1["Global_Sales"].sum()

#index1=vgdata["Rating"].count()
#Create Dataset of ratings
yearly_ratings=pd.DataFrame({"Global Sales of E Rated Games by year": rate_E,
                     "Global Sales of E10+ Rated Games by year": rate_Eten,
                    "Global Sales of T Rated Games by year": rate_T,
                     "Global Sales of M Rated Games by year": rate_M})
#yearly_ratings.set_index("Rating")

yearly_ratings.head()
```

```python
#Plot Ratings by year by line plot
yearly_ratings.plot(kind="line", figsize= (20,10))
plt.savefig("Ratings_Graph")
plt.show()
```
